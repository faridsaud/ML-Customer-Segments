```python
# Investigate "CAMEO_INTL_2015" and engineer two new variables.

def add_variables(df):
    variable_name = "CAMEO_INTL_2015"
    values_wealth = []
    values_life_stage = []
    for index, row in df.iterrows():
        if pd.isnull(row[variable_name]) or row[variable_name] == 'XX':
            values_wealth.append(np.nan)
            values_life_stage.append(np.nan)
        else:
            # Add value for the wealth column
            number = int(int(row[variable_name])/10)
            if number > 0:
                values_wealth.append(number)
            else:
                values_wealth.append(np.nan)
                
            # Add value for the life_stage column
            number = int(row[variable_name]) % 10
            if number > 0:
                values_life_stage.append(number)
            else:
                values_life_stage.append(np.nan)
        if index % 100000 == 0:
            print(index)
    df["CAMEO_INTL_2015_WEALTH"] = values_wealth
    df["CAMEO_INTL_2015_LIFE_STAGE"] = values_life_stage
    return df


add_variables(reeconded_df)

print(reeconded_df.shape)
print(reencoded_dataframe.shape)
```

```python
def convert_missing_values(df):
    return df.replace(to_replace={-1: np.nan, 0: np.nan})


# COLUMNS

def get_columns_removed(df_1, df_2):
    df_1_columns = df_1.columns.values
    df_2_columns = df_2.columns.values
    removed_columns = []
    for column in df_1_columns:
        if not column in df_2_columns:
            removed_columns.append(column)
    return removed_columns

def remove_columns(df, columns_to_remove = None):
    if not columns_to_remove is None:
        df_without_nan = df.drop(columns_to_remove, axis=1)
        return df_without_nan, columns_to_remove
    remove_ratio = 0.5
    total_rows = df.shape[0]
    min_existing_values_per_column = int((1-remove_ratio) * total_rows)
    df_without_nan = df.dropna(thresh=min_existing_values_per_column, axis=1)
    removed_columns = get_columns_removed(df, df_without_nan)
    return df_without_nan, removed_columns

# ROWS
def remove_rows(df):
    remove_ratio = 0.5
    total_columns = df.shape[1]
    min_existing_values_per_row = int((1-remove_ratio) * total_columns)
    return df.dropna(thresh=min_existing_values_per_row)
    

# ENCODING    
def get_categorical_variables(info_df, df):
    categorical_variables = {}
    headers = list(df.columns.values)
    for index, row in info_df.iterrows():
        if row['attribute'] in headers:
            if row['type'] in categorical_variables:
                categorical_variables[row['type']].append(row['attribute'])
            else:
                categorical_variables[row['type']] = []
    return categorical_variables


def reencode_multilevel_variables(df, features):
    return pd.get_dummies(df, columns=features)

# ENCODING - PRAEGENDE_JUGENDJAHRE
def convert_column_1(df):
    variable = {
        'name': "PRAEGENDE_JUGENDJAHRE",
        'new_variables': ["PRAEGENDE_JUGENDJAHRE_MOVEMENT", "PRAEGENDE_JUGENDJAHRE_DECADE"],
        'maps': [{
            '1': 'Mainstream',
            '2': 'Avantgarde',
            '3': 'Mainstream',
            '4': 'Avantgarde',
            '5': 'Mainstream',
            '6': 'Avantgarde',
            '7': 'Avantgarde',
            '8': 'Mainstream',
            '9': 'Avantgarde',
            '10': 'Mainstream',
            '11': 'Avantgarde',
            '12': 'Mainstream',
            '13': 'Avantgarde',
            '14': 'Mainstream',
            '15': 'Avantgarde',
        }, {
            '1': 1,
            '2': 1,
            '3': 2,
            '4': 2,
            '5': 3,
            '6': 3,
            '7': 3,
            '8': 4,
            '9': 4,
            '10': 5,
            '11': 5,
            '12': 5,
            '13': 5,
            '14': 6,
            '15': 6,
        }]
    }
    def add_variable(base_variable_name, variable_name, variable_map, df):
        values = []
        for index, row in df.iterrows():
            if np.isnan(row[base_variable_name]):
                values.append(np.nan)
            else:
                values.append(variable_map[str(int(row[base_variable_name]))])
        df[variable_name] = values
        return df
    add_variable(variable['name'], variable['new_variables'][0],  variable['maps'][0], df)
    add_variable(variable['name'], variable['new_variables'][1],  variable['maps'][1], df)
    df = pd.get_dummies(df, columns=[variable['new_variables'][0]])
    return df.drop(variable['name'], 1)

# ENCODING - CAMEO_INTL_2015
def convert_column_2(df):
    def add_variables(df):
        variable_name = "CAMEO_INTL_2015"
        values_wealth = []
        values_life_stage = []
        for index, row in df.iterrows():
            if pd.isnull(row[variable_name]) or row[variable_name] == 'XX':
                values_wealth.append(np.nan)
                values_life_stage.append(np.nan)
            else:
                # Add value for the wealth column
                number = int(int(row[variable_name])/10)
                if number > 0:
                    values_wealth.append(number)
                else:
                    values_wealth.append(np.nan)

                # Add value for the life_stage column
                number = int(row[variable_name]) % 10
                if number > 0:
                    values_life_stage.append(number)
                else:
                    values_life_stage.append(np.nan)
        df["CAMEO_INTL_2015_WEALTH"] = values_wealth
        df["CAMEO_INTL_2015_LIFE_STAGE"] = values_life_stage
        return df
    df = add_variables(df)
    df = df.drop('CAMEO_INTL_2015', 1)
    return df

def convert_mixed_variables(df):
    df = convert_column_1(df)
    df = convert_column_2(df)
    return df

def transform_features(df, feat_df):
    categorical_variables = get_categorical_variables(feat_df, df)
    df = reencode_multilevel_variables(df, categorical_variables['categorical'])
    df = convert_mixed_variables(df)
    return df

def clean_data(df, info_df, columns_to_remove = None):
    """
    Perform feature trimming, re-encoding, and engineering for demographics
    data
    
    INPUT: Demographics DataFrame, Info DataFrame
    OUTPUT: Trimmed and cleaned demographics DataFrame
    """
    df = df.copy()
    info_df = info_df.copy()
    
    # Put in code here to execute all main cleaning steps:
    # convert missing value codes into NaNs, ...
    print('cleaning missing values')
    df = convert_missing_values(df)
    print(df.shape)
    
    # remove selected columns and rows, ...
    print('removing rows and cols')
    df, columns_removed = remove_columns(df, columns_to_remove)
    print(df.shape)
    df = remove_rows(df)
    print(df.shape)

    
    # select, re-encode, and engineer column values.
    print('reencoding')
    df = transform_features(df, info_df)
    print(df.shape)
    # Return the cleaned dataframe.
    return df, columns_removed

cleaned_df, columns_removed = clean_data(azdias, feat_info)

# Tests
print(cleaned_df['PRAEGENDE_JUGENDJAHRE_MOVEMENT_Mainstream'][1] == 1)
print(cleaned_df['PRAEGENDE_JUGENDJAHRE_MOVEMENT_Avantgarde'][2] == 1)
print(cleaned_df['PRAEGENDE_JUGENDJAHRE_DECADE'][1] == 6)
print(cleaned_df['PRAEGENDE_JUGENDJAHRE_DECADE'][2] == 6)
print(cleaned_df['CAMEO_INTL_2015_WEALTH'][1] * 10 + cleaned_df['CAMEO_INTL_2015_LIFE_STAGE'][1] == float(azdias['CAMEO_INTL_2015'][1]))
print(cleaned_df['CAMEO_INTL_2015_WEALTH'][10] * 10 + cleaned_df['CAMEO_INTL_2015_LIFE_STAGE'][10] == float(azdias['CAMEO_INTL_2015'][10]))
print(cleaned_df.shape[0] == 817177)
print(cleaned_df.shape[1] == 195)
```

To start, you should not set any parameters (so all components are computed) or set a number of components that is at least half the number of features (so there's enough features to see the general trend in variability).

```python
# Apply PCA to the data.
def do_pca(data, n_components = None):
    '''
    Transforms data using PCA to create n_components, and provides back the results of the
    transformation.
    
    INPUT: n_components - int - the number of principal components to create
           data - the data you would like to transform
           
    OUTPUT: pca - the pca object created after fitting the data
            X_pca - the transformed X matrix with new number of components
    '''
    pca = PCA(n_components)
    X_pca = pca.fit_transform(data)
    return pca, X_pca

pca, X_pca = do_pca(dataset)
```

```python
# Investigate the variance accounted for by each principal component.

def plot_pca(pca, min_variance = 0.85):
    '''
    Creates a scree plot associated with the principal components 
    
    INPUT: pca - the result of instantian of PCA in scikit learn
            
    OUTPUT:
            None
    '''
    num_components = len(pca.explained_variance_ratio_)
    ind = np.arange(num_components)
    vals = pca.explained_variance_ratio_
 
    plt.figure(figsize=(10, 6))
    ax = plt.subplot(111)
    cumvals = np.cumsum(vals)
    ax.bar(ind, vals)
    ax.plot(ind, cumvals)
    breakpoint_printed = False
    for i in range(num_components):
        ax.annotate(r"%s%%" % ((str(vals[i]*100)[:4])), (ind[i]+0.2, vals[i]), va="bottom", ha="center", fontsize=12)
        if cumvals[i] > min_variance and not breakpoint_printed:
            ax.scatter([i], [cumvals[i]])
            ax.annotate('N = {n} , TOTAL VARIANCE = {cum_variance}'.format(n=i, cum_variance=cumvals[i]), (i, cumvals[i]))
            breakpoint_printed = True
 
    ax.xaxis.set_tick_params(width=0)
    ax.yaxis.set_tick_params(width=2, length=12) 
    ax.set_xlabel("Principal Component")
    ax.set_ylabel("Variance Explained (%)")
    plt.title('Explained Variance Per Principal Component')
    
    
plot_pca(pca)
print(X_pca.shape)
```

### Step 2.3: Interpret Principal Components

Now that we have our transformed principal components, it's a nice idea to check out the weight of each variable on the first few components to see if they can be interpreted in some fashion.

As a reminder, each principal component is a unit vector that points in the direction of highest variance (after accounting for the variance captured by earlier principal components). The further a weight is from zero, the more the principal component is in the direction of the corresponding feature. If two features have large weights of the same sign (both positive or both negative), then increases in one tend expect to be associated with increases in the other. To contrast, features with different signs can be expected to show a negative correlation: increases in one variable should result in a decrease in the other.

- To investigate the features, you should map each weight to their corresponding feature name, then sort the features according to weight. The most interesting features for each principal component, then, will be those at the beginning and end of the sorted list. Use the data dictionary document to help you understand these most prominent features, their relationships, and what a positive or negative value on the principal component might indicate.

To help facilitate this, you should write a function that you can call at any time to print the sorted list of feature weights, for the *i*-th principal component. This might come in handy in the next step of the project, when you interpret the tendencies of the discovered clusters.

```python
# Map weights for the first principal component to corresponding feature names
# and then print the linked values, sorted by weight.
# HINT: Try defining a function here or in a new cell that you can reuse in the
# other cells.


def get_weight_map(pca, component_index, dataframe):
    weight_dict = {}
    columns = dataframe.columns
    principal_component = pca.components_[component_index]
    for i in  range(len(principal_component)):
        weight = principal_component[i]
        weight_dict[columns[i]] = weight
    return sorted(weight_dict.items(), key=lambda x:x[1], reverse=True)

get_weight_map(pca, 0, cleaned_df)

```
